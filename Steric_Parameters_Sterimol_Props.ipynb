#**Sterimol**

```python
# Don't touch this cell except for the first run
all = pd.DataFrame()

def letter_to_int(letter):
  alphabet = list('abcdefghijklmnopqrstuvwxyz')
  return alphabet.index(letter)

# Glossary for atom numbers and compound names in alphabetical order
glossary = pd.read_csv('https://github.com/xh313/compchem/raw/main/sfamines_an%20(1).csv')
glossary
```

```python
#CHANGE THE DIRECTORY NAME to identify where your xyz files are
#Most common error can be fixed by putting an r before the directory path
#Check to be sure all files in folder are .xyz (write excel files to a separate folder)

num = letter_to_int('a')  # molecule code, change this

directory = f'/content/drive/MyDrive/xyzfiles/{glossary.molecule[num]}'

#ONLY OTHER THINGS TO CHANGE ARE ATOM1 AND ATOM2 NUMBERS
#atom1 is base atom and atom2 is first atom of substituent looking down bond in question
an1 = glossary.S[num]
an2 = glossary['linked atom'][num]

print(f'S: {an1}; Linked atom:{an2}')

files = os.listdir(directory)
files = sorted(files)
print('Files:', len(files))
print(files)
```

```python
#Writes an empty data frame (df) to be filled in with values as they are computed
#DO NOT EVER NEED TO CHANGE THIS
params = {'File':[], 'L':[], 'Bmin':[], 'Bmax':[]}
df = pd.DataFrame(params, columns = ['File','L','Bmin','Bmax'])
df
```

```python
#ALWAYS USE GRID MODE AND DO NOT CHANGE OTHER SETTINGS!
count = 0
#directory = r'C:\Users\rcamm\Downloads\Theo_testing_forBV'
for i in files:
    fp = str(directory)+'/'+str(i)
    print(fp)
    mol = db.dbstep(fp,atom1=an1,atom2=an2,commandline=True,
                    verbose=True,sterimol=True,measure='grid')
    L = mol.L
    Bmin = mol.Bmin
    Bmax = mol.Bmax
    x = [i,L,Bmin,Bmax]
    df.loc[count] = x
    count+=1
```

```python
#Writes output values to the data frame
df
```

```python
# Combine all output lines into this file
# Run this code cell if you wish to append the current result in the sheet

all = pd.concat([all, df])
all
```

```python
#Exports ALL results to an excel spreadsheet
#WILL OVERWRITE IF RE-RUN AND RE-ADD TO DATA FRAME BEFORE EXPORTING TO EXCEL
all.to_excel(f'/content/Sterimol_all.xlsx', index = False)
```

#**Buried volume**

```python
# Run this before anything
!pip install dbstep
import dbstep.Dbstep as db
import pandas as pd
import os
import openpyxl
#import openbabel
print('done')
```

```python
# Don't touch this cell except for the first run
all = pd.DataFrame()
glossary = pd.read_csv('https://github.com/xh313/compchem/raw/main/sfamines_an%20(1).csv')
glossary
```

```python
# THE ONLY CELL YOU NEED TO CHANGE STUFF IN!
#CHANGE THE DIRECTORY NAME to identify where your xyz files are
#Most common error can be fixed by putting an r before the directory path
#Check to be sure all files in folder are .xyz 
#(write excel files to a separate folder)

def letter_to_int(letter):
  alphabet = list('abcdefghijklmnopqrstuvwxyz')
  return alphabet.index(letter)

num = letter_to_int('u')  # molecule code, change this

directory = f'/content/xyzfiles/{glossary.molecule[num]}'  # change this

an = glossary.S[num]  # atom number of the atom of interest, change this accordingly

print(an)
```

```python
# Check list of files
files = os.listdir(directory)
files = sorted(files)
print(len(files))
print(files)
```

```python
#Writes an empty data frame (df) to be filled in with values as they are computed
#DO NOT EVER NEED TO CHANGE THIS
params = {'File':[], 'bur_vol_r225':[], 'bur_vol_r250':[], 'bur_vol_r275':[], 'bur_vol_r300':[]}
df = pd.DataFrame(params, columns = ['File','bur_vol_r225','bur_vol_r250','bur_vol_r275','bur_vol_r300'])
df
```

```python
#ONLY OTHER THINGS TO CHANGE ARE ATOM1 NUMBER AND SPHERE RADIUS (r)
#Be sure that mol# matches mol# and bur_vol_label 
#matches data frame column labels above
#ALWAYS USE GRID MODE AND DO NOT CHANGE OTHER SETTINGS!
count = 0

for i in files:
    fp = str(directory)+'/'+str(i)
    print(fp)
    mol1 = db.dbstep(fp,atom1=an,commandline=True,verbose=True,
                     volume=True,r=2.25,measure='grid')
    bur_vol_r225 = mol1.bur_vol 
    mol2 = db.dbstep(fp,atom1=an,commandline=True,verbose=True,
                     volume=True,r=2.5,measure='grid')
    bur_vol_r250 = mol2.bur_vol  
    mol3 = db.dbstep(fp,atom1=an,commandline=True,verbose=True,
                     volume=True,r=2.75,measure='grid')
    bur_vol_r275 = mol3.bur_vol 
    mol4 = db.dbstep(fp,atom1=an,commandline=True,verbose=True,
                     volume=True,r=3.0,measure='grid')
    bur_vol_r300 = mol4.bur_vol
    x = [i,bur_vol_r225,bur_vol_r250,bur_vol_r275,bur_vol_r300]
    df.loc[count] = x
    count+=1
```

```python
#Writes output values to the data frame
df
```

```python
# Combine all output lines into this file
# Run this code cell if you wish to append the current result in the sheet

all = pd.concat([all, df])
all
```

```python
#Exports CURRENT results to an excel spreadsheet
#WILL OVERWRITE IF RE-RUN AND RE-ADD TO DATA FRAME BEFORE EXPORTING TO EXCEL
# Excel file named according to the file name of your xyz file

df.to_excel(f'/content/BV_test_{files}.xlsx', index = False)
```

```python
#Exports ALL results to an excel spreadsheet
#WILL OVERWRITE IF RE-RUN AND RE-ADD TO DATA FRAME BEFORE EXPORTING TO EXCEL
all.to_excel(f'/content/BV_test_all.xlsx', index = False)
```
